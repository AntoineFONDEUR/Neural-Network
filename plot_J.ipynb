```python
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import clear_output
import time
```

```python
while True:
    l=[]
    first_line = True
    
    with open('log.txt', 'r') as file:
        for line in file:
            if first_line:
                learning_rate, epochs, batch_size, lamb = line.strip().split(',')
                first_line=False
            else: l.append(float(line.strip()))
    
    learning_rate = float(learning_rate)
    epochs = float(epochs)
    batch_size = float(batch_size)
    lamb = float(lamb)
    J = np.array(l)
    
    clear_output(wait=True)
    plt.title(f"J per iteration for alpha = {learning_rate}, epochs = {epochs}, batch size = {batch_size} and lambda = {lamb} \n")
    plt.plot(J)
    plt.show()
    if input('>>>') == 'q': break

plt.figure(figsize=(10,5))
plt.title(f"J per iteration for alpha = {learning_rate}, epochs = {epochs}, batch size = {batch_size} and lambda = {lamb} \n")
plt.plot(J)
text = str(learning_rate).replace(".",",") + '-' + str(int(epochs)) + '-' + str(int(batch_size)) + '-' + str(lamb).replace(".",",")
plt.savefig("plots/" + text +".png")
```
